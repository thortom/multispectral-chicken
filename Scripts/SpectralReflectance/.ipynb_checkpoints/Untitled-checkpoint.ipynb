```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import os

data = None

# Read all file in foled
for fileName in os.listdir():
    if '.csv' in fileName:
        print(fileName)

        colName = fileName[16:-4]
        tmp = pd.read_csv(fileName, sep=";", index_col=0, names=[colName])

        # Merge the file to one bigger array with correctly labeld coloms
        if data is None:
            data = tmp
        else:
            data[colName] = tmp[colName]

print(data.head())
print(data.info())
```

```python
data.plot()
plt.show()
```

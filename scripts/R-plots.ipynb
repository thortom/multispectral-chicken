```r
install.packages("ggfortify")
```

```r
library(ggfortify)
df <- iris[c(1, 2, 3, 4)]
autoplot(prcomp(df))
```

```r
autoplot(prcomp(df), data = iris, colour = 'Species')
```

```r
autoplot(prcomp(df), data = iris, colour = 'Species',
         loadings = TRUE, loadings.colour = 'blue',
         loadings.label = TRUE, loadings.label.size = 3)
```
